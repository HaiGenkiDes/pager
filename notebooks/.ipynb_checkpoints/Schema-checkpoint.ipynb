```python
#stdlib
import os.path

#third party
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from losspager.schema import emailschema as es
from shapely.geometry import asShape, shape
```

Load the anonymized database.  The names, email addresses, and organizations have all been set to combinations of random words selected from the Unix "words" file.

```python
dbfile = os.path.join(os.getcwd(),'..','test','data','anonymized_pager_users.db')
url = es.get_file_url(dbfile)
session = es.get_session(url)
```

Perform some simple queries to get the number of users and number of organizations.

```python
print('There are %i users in the database.' % session.query(es.User).count())
print('There are %i organizations in the database.' % session.query(es.Organization).count())
```

Print out the first ten organizations in the database - remember, these have been constructed with *random* words!

```python
for org in session.query(es.Organization).limit(80):
    print(org)
```

Print out a list of the first 5 users, their addresses, and then up to three of the versions linked to each of those addresses.

```python
for user in session.query(es.User).limit(5):
    print(user)
    for address in user.addresses:
        print('\t'+str(address))
        nv = 0
        for version in address.versions:
            print('\t\t'+str(version))
            nv += 1
            if nv > 2: #there are a LOT of versions associated with each address...
                break
```

Make maps of each of the regions in each region group.

```python
tpoly = None
for regiongroup in session.query(es.RegionGroup):
    print(regiongroup)
    fig=plt.figure(figsize=(8,4.5))
    ax = fig.add_axes([0.05,0.05,0.9,0.85])
    #m = Basemap(llcrnrlon=0,llcrnrlat=-80,urcrnrlon=360,urcrnrlat=80,projection='mill')
    m = Basemap(projection='hammer',lon_0=180)
    plt.title('Regions for group %s' % regiongroup.groupname)
    for region in regiongroup.regions:
        print('\t'+str(region))
        poly = region.getPolygon()
        lon,lat = poly.exterior.xy
        x,y = m(lon,lat)
        m.plot(x,y)
        m.drawcoastlines()
        

        
```
